```python
# Load data
hsc_avg = pd.read_excel('./Datasets/compiled.xlsx', sheet_name="HSC_Avg_2").set_index("Unnamed: 0")

# Plot heatmap
plt.figure(figsize=(12, 10))
heatmap = sns.heatmap(
    hsc_avg, 
    annot=True, 
    cmap="coolwarm", 
    fmt=".3f", 
    linewidths=0.5,
    annot_kws={"size": 18},
    cbar_kws={'label': 'MAPE'}
)

# Set colorbar label size
colorbar = heatmap.collections[0].colorbar
colorbar.ax.yaxis.label.set_size(18)  # Set colorbar label size

# Set title and axis font sizes
plt.title("Performance Heatmap for Average HSC MAPE", fontsize=18)
plt.xlabel("Upsampling Method", fontsize=18)
plt.ylabel("Model Type", fontsize=18)
plt.xticks(fontsize=18, rotation=45, ha='right')
plt.yticks(fontsize=18)

plt.tight_layout()
plt.show()
```

```python
qaly_over_n = pd.read_excel('./Datasets/compiled.xlsx', sheet_name="QALY_Over_n")
# Plot line graph
plt.figure(figsize=(8, 6))
sns.lineplot(x=qaly_over_n["n"], y=qaly_over_n["Avg_MAPE"], marker="o", linestyle="-", color="b")

plt.xlabel("Oversampling (n)")
plt.ylabel("Average MAPE")
plt.title("Effect of Oversampling Number of Samples on QALY MAPE")
plt.grid(True)

plt.show()
```

```python

hsc_over_n = pd.read_excel('./Datasets/compiled.xlsx', sheet_name="HSC_Over_n")
# Plot line graph
plt.figure(figsize=(8, 6))
sns.lineplot(x=hsc_over_n["n"], y=hsc_over_n["Avg_MAPE"], marker="o", linestyle="-", color="b")

plt.xlabel("Oversampling (n)", fontsize=18)
plt.ylabel("Average MAPE", fontsize=18)
plt.title("Effect of Upsampling Level on HSC MAPE", fontsize=18)
plt.xticks(fontsize=18)
plt.yticks(fontsize=18)
plt.grid(True)

plt.show()
```

```python
qaly_boot_avg = pd.read_excel('./Datasets/compiled.xlsx', sheet_name="QALY_Bootstrap_Avg")
# Plotting the line graph
plt.figure(figsize=(8, 5))
plt.plot(qaly_boot_avg['n'], qaly_boot_avg['No Bootstrap'], marker='o', label='No Bootstrap')
plt.plot(qaly_boot_avg['n'], qaly_boot_avg['Bootstrap'], marker='o', label='Bootstrap')
plt.xlabel('Oversampling Level (n)')
plt.ylabel('MAPE')
plt.title('Comparison of Bootstrap and No Bootstrap Across Oversampling Levels')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
