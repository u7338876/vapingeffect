```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
qaly_50 = pd.read_excel('./Datasets/compiled.xlsx', sheet_name="QALY_n=50").set_index("Unnamed: 0")

# Plot heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(qaly_50, annot=True, cmap="coolwarm", fmt=".3f", linewidths=0.5)

plt.xticks(rotation=0)

plt.title("Performance Heatmap for QALY MAPE when Oversampling is n=50")
plt.show()
```
